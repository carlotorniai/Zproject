```python
# Now let's filter my dataframe to keep only the information for this data
reduced_df = df_skills[skills_count_30]
```

```python
reduced_df
```

```python
# Now If I want to create a hierarchical clustering let's do so
# Let's do the clusterign with this new dataframe now
dist_skills = squareform(pdist(reduced_df, metric='euclidean'))
```

```python
dist_skills.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pymongo
fig = plt.figure(figsize=(60, 20))
# I want to use as the index the job title to get some information
#for i in range(len(reduced_df.index)):
#    print reduced_df.index[i]

R = dendrogram(linkage(dist_skills, method='complete'), labels = reduced_df.index, color_threshold=22, leaf_font_size=10)
```

```python
# Let's build a consistent set of labels that reports in addition to the User_id the current job title 
import pdb
new_lables = []
# Connect to DB
db, collection = utils.initializeDb("zproject", "ext_profiles_processed")
for i in range(len(reduced_df.index)):
    # print reduced_df.index[i]
    user_id = reduced_df.index[i]
    # Here thet the current title for the position if available
    cursor = collection.find({"id":user_id, "positions.values.isCurrent" : True}, {"_id" : 0, "positions.values.title" : 1})
    new_label=""
    for position in cursor:
         title =  position['positions']['values'][0]['title']
         new_label = title
    new_lables.append(new_label)
    
fig = plt.figure(figsize=(60, 20))
# I want to use as the index the job title to get some information
#for i in range(len(reduced_df.index)):
#    print reduced_df.index[i]

R = dendrogram(linkage(dist_skills, method='complete'), labels = new_lables, color_threshold=22, leaf_font_size=10)
```
